```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
import math
```

### –î–æ–º–∞—à–Ω—ñ –∑–∞–≤–¥–∞–Ω–Ω—è –Ω–∞ –≤—ñ–¥–ø—Ä–∞—Ü—é–≤–∞–Ω–Ω—è –ø–æ–Ω—è—Ç—Ç—è —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó

1. –ó–∞–¥–∞–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è –∑–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ –æ—Ü—ñ–Ω–∫–∏ –≤—ñ–¥ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≥–æ–¥–∏–Ω, –≤–∏—Ç—Ä–∞—á–µ–Ω–∏—Ö –Ω–∞ –Ω–∞–≤—á–∞–Ω–Ω—è: $ f(x) = 10x - 5 $.
  1. –ó –¥–æ–ø–æ–º–æ–≥–æ—é Python –æ–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—É —Ü—ñ—î—ó —Ñ—É–Ω–∫—Ü—ñ—ó.
  2. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –±—ñ–±–ª—ñ–æ—Ç–µ–∫—É `matplotlib` –¥–ª—è –ø–æ–±—É–¥–æ–≤–∏ –≥—Ä–∞—Ñ—ñ–∫—ñ–≤ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –∑–Ω–∞—á–µ–Ω—å –≤—ñ–¥ 0 –¥–æ 10 –≥–æ–¥–∏–Ω. –ü—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–π—Ç–µ, —è–∫ –∑–º—ñ–Ω–∞ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≥–æ–¥–∏–Ω –≤–∏–≤—á–µ–Ω–Ω—è –≤–ø–ª–∏–≤–∞—î –Ω–∞ –æ—Ü—ñ–Ω–∫—É.



```python
# –§—É–Ω–∫—Ü—ñ—è ùëì(ùë•)=10ùë•‚àí5
def f(x):
  return 10 * x - 5
```

```python
x = np.linspace(0, 10, 10)
```

```python
# –ü–æ—Ö—ñ–¥–Ω–∞ f'(x) = 2x
df = 2 * x
```

```python
# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –≥—Ä–∞—Ñ—ñ–∫—É
plt.figure(figsize=(15, 5))
plt.plot(x, f(x), label='f(x)=10x‚àí5', color='blue')
plt.plot(x, df, label="f'(x)=2x", color='orange')

# –î–æ–¥–∞–≤–∞–Ω–Ω—è –ª–µ–≥–µ–Ω–¥–∏, –º—ñ—Ç–æ–∫ —Ç–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞
plt.title("–ì—Ä–∞—Ñ—ñ–∫ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show();
```

**–í–∏—Å–Ω–æ–≤–æ–∫:**
- f(x)  —î –ª—ñ–Ω—ñ–π–Ω–æ—é —ñ –∑—Ä–æ—Å—Ç–∞—î –∑ —á–∞—Å–æ–º. –¶–µ –≥–æ–≤–æ—Ä–∏—Ç—å –ø—Ä–æ —Ç–µ, —â–æ —á–∏–º –±—ñ–ª—å—à–µ —á–∞—Å—É –º–∏ –Ω–∞–≤—á–∞—î–º–æ—Å—å, —Ç–∏–º –∫—Ä–∞—â–∞ –æ—Ü—ñ–Ω–∫–∞
- f'(x) –∑–∞–≤–∂–¥–∏ –º–∞—î –æ–¥–Ω–∞–∫–æ–≤–µ –∑–Ω–∞—á–µ–Ω–Ω—è (2,22) —Ü–µ –≥–æ–≤–æ—Ä–∏—Ç—å –ø—Ä–æ —Ç–µ, —â–æ –ø—Ä–∏—Ä—ñ—Å—Ç –æ—Ü—ñ–Ω–∫–∏ –∑–∞ –∫–æ–∂–Ω—É –≥–æ–¥–∏–Ω—É –æ–¥–Ω–∞–∫–æ–≤–∏–π —ñ –Ω–µ –∑–∞–ª–µ–∂–∏—Ç—å –≤—ñ–¥ —Ç–æ–≥–æ —Å–∫—ñ–ª—å–∫–∏ –≤–∏ –≤–∂–µ –Ω–∞–≤—á–∏–ª–∏—Å—è
- –æ—Ç–∂–µ –º–æ–∂–µ–º–æ –∑—Ä–æ–±–∏—Ç–∏ –≤–∏—Å–Ω–æ–≤–æ–∫, —â–æ –∫—ñ–ª-—Ç—å –≥–æ–¥–∏–Ω –≤–∏—Ç—Ä–∞—á–µ–Ω–∏—Ö –Ω–∞ –Ω–∞–≤—á–∞–Ω–Ω—è –ø–æ–∑–∏—Ç–∏–≤–Ω–æ –≤–ø–ª–∏–≤–∞—é—Ç—å –Ω–∞ –æ—Ü—ñ–Ω–∫—É, –∞–ª–µ –∑ —á–∞—Å–æ–º —Ü–µ–π —Ç–µ–º —É–ø–æ–≤—ñ–ª—å–Ω—é—î—Ç—å—Å—è

2. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ Python —ñ `SymPy`, –æ–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—ñ –¥–ª—è —Ñ—É–Ω–∫—Ü—ñ–π $ f(x) = \sin(x) $ —Ç–∞ $ f(x) = \ln(x) $. –ó–æ–±—Ä–∞–∑—ñ—Ç—å —Ü—ñ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—Ö–Ω—ñ –ø–æ—Ö—ñ–¥–Ω—ñ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é `matplotlib`.



```python
# –í–∏–∑–Ω–∞—á–∞—î–º–æ —Å–∏–º–≤–æ–ª—ñ—á–Ω—É –∑–º—ñ–Ω–Ω—É
x = sp.symbols('x')

# –í–∏–∑–Ω–∞—á–∞—î–º–æ —Ñ—É–Ω–∫—Ü—ñ–π
f_sin = sp.sin(x)
f_ln = sp.log(x)

# –í–∏–∑–Ω–∞—á–∞—î–º–æ –ø–æ—Ö—ñ–¥–Ω—ñ —Ñ—É–Ω–∫—Ü—ñ–π
df_sin = sp.diff(f_sin, x)
df_ln = sp.diff(f_ln, x)

print("–ü–æ—Ö—ñ–¥–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—ó ùëì(ùë•)=sin(ùë•):", df_sin)
print("–ü–æ—Ö—ñ–¥–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—ó ùëì(ùë•)=ln(ùë•):", df_ln)
```

```python
x_val = np.linspace(1, 20, 50)

out_x_val_sin = [] # —Ñ—É–Ω–∫—Ü—ñ—è ùëì(ùë•)=sin(ùë•)
for i in range(len(x_val)): 
    out_x_val_sin.append(math.sin(x_val[i])) 
    i += 1


out_x_val_cos = [] # –ü–æ—Ö—ñ–¥–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—ó ùëì(ùë•)=sin(ùë•): cos(x)
for i in range(len(x_val)): 
    out_x_val_cos.append(math.cos(x_val[i])) 
    i += 1


out_x_val_ln = [np.log(x) for x in x_val] # —Ñ—É–Ω–∫—Ü—ñ—è ùëì(ùë•)=ln(ùë•)

out_x_val=1 / x_val


# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –≥—Ä–∞—Ñ—ñ–∫—É
plt.figure(figsize=(15, 5))
plt.plot(x_val, out_x_val_sin, label='f(x)=sin(x)', color='blue')
plt.plot(x_val, out_x_val_cos, label="f'(x)=cos(x)", color='blue', linestyle='dashed')
plt.plot(x_val, out_x_val_ln, label="f(x)=log(x)", color='orange')
plt.plot(x_val, out_x_val, label="f'(x)=1/x", color='orange', linestyle='dashed')

# –î–æ–¥–∞–≤–∞–Ω–Ω—è –ª–µ–≥–µ–Ω–¥–∏, –º—ñ—Ç–æ–∫ —Ç–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞
plt.title("–ì—Ä–∞—Ñ—ñ–∫ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show();
```


3. –ó–∞–≥–∞–ª—å–Ω—ñ –≤–∏—Ç—Ä–∞—Ç–∏ –∫–æ–º–ø–∞–Ω—ñ—ó –≤—ñ–¥ –æ–±—Å—è–≥—É –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–∞ –∑–∞–¥–∞–Ω—ñ —Ñ—É–Ω–∫—Ü—ñ—î—é $ f(x) = x^2 - 5x + 1 $, –¥–µ $ x $ ‚Äî –∫—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∏—Ä–æ–±–ª–µ–Ω–∏—Ö –æ–¥–∏–Ω–∏—Ü—å. –û–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—É —Ü—ñ—î—ó —Ñ—É–Ω–∫—Ü—ñ—ó –∑ –¥–æ–ø–æ–º–æ–≥–æ—é SymPy —Ç–∞ –≤—ñ–∑—É–∞–ª—ñ–∑—É–π—Ç–µ –∑–Ω–∞—á–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó —ñ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó –Ω–∞ –≤—ñ–¥—Ä—ñ–∑–∫—É –≤—ñ–¥ 0 –¥–æ 100.

```python
# –í–∏–∑–Ω–∞—á–∞—î–º–æ —Å–∏–º–≤–æ–ª—ñ—á–Ω—É –∑–º—ñ–Ω–Ω—É
x = sp.symbols('x')

f = x ** 2 - 5 * x + 1 # –§—É–Ω–∫—Ü—ñ—è ùëì(ùë•)=ùë•2‚àí5ùë•+1
d_f = sp.diff(f, x) # –ø–æ—Ö—ñ–¥–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—ó ùëì(ùë•)=ùë•2‚àí5ùë•+1

print("–ü–æ—Ö—ñ–¥–Ω–∞ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ—ó —Ñ—É–Ω–∫—Ü—ñ—ó x^2 + 5x + 1:", d_f)
```

```python
x = np.linspace(0, 100, 10)

f = x ** 2 - 5 * x + 1 # –§—É–Ω–∫—Ü—ñ—è ùëì(ùë•)=ùë•2‚àí5ùë•+1
d_f = 2 * x - 5 # –ø–æ—Ö—ñ–¥–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—ó ùëì(ùë•)=ùë•2‚àí5ùë•+1


# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –≥—Ä–∞—Ñ—ñ–∫—É
plt.figure(figsize=(15, 5))
plt.plot(x, f, label='ùëì(x)=x^2‚àí5x+1', color='blue')
plt.plot(x, d_f, label="f'(x)=2x - 5", color='blue', linestyle='dashed')


# –î–æ–¥–∞–≤–∞–Ω–Ω—è –ª–µ–≥–µ–Ω–¥–∏, –º—ñ—Ç–æ–∫ —Ç–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞
plt.title("–ì—Ä–∞—Ñ—ñ–∫ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show();
```
